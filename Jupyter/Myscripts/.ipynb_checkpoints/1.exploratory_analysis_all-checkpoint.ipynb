# Look at some basic qc information about the experiments

```python
qcs = iglob('../Mydata/qc_metrics/*.txt')

dfs = []
for qc in qcs:
    dfs.append(pd.read_table(qc, index_col=0))
    
plate_qc = pd.concat(dfs, axis=1, sort=True)
plate_qc['Cell'] = plate_qc.index.str.extract('(.*)-.*', expand=False)

sort_idx = plate_qc.index.sort_values()
plate_qc =  plate_qc.loc[sort_idx]

plate_qc.head()
```

```python
# look at the isize distritubtion
fig, ax = plt.subplots(figsize=(5,5))

# read the isize of aggregated data
merged = pd.read_table('../Mydata/Merged_all_gmd_isize.hist',
                       sep=' ', header=None, names=['count', 'size'])

# read 24 single cells to plot
examples = ['../Mydata/8-10cell/8-10cell-{:0>1}.hist'\
            .format(i) for i in range(56,71)]

scs = []
for i in examples:
    scs.append(pd.read_table(i, sep=' ', header=None, names=['count', 'size']))

ax.plot(scs[0]['size'],
        scs[0]['count']/scs[0]['count'].sum() * 100,
        'k-', label='Single cell examples', lw=.5, alpha=.1)

for sc in scs[1:]:
    ax.plot(sc['size'],
            sc['count']/sc['count'].sum() * 100,
            'k-', lw=.5, alpha=.1, label='')

ax.plot(merged['size'],
        merged['count']/merged['count'].sum() * 100,
        'r-', lw=1.5, label='Aggregate of all single cells')

lg = ax.legend()
lg_lines = lg.get_lines()
lg_lines[0].set_linewidth(2)
lg_lines[0].set_alpha(.5)

ax.set_xlim(0,1200)
ax.set_xlabel('Insert size (bp)')
ax.set_ylabel('Frequency (%)', )

plt.savefig('../Myfigure/fragment_size_distribution.pdf',
             bbox_inches='tight', transparent=True)
```

# Perform some basic exploratory analysis

```python
from scipy.io import mmread

count = mmread('../Mydata/Count_matrix/mm_scATAC_count_matrix_over_all.mtx')
idxs = [i.strip() for i in open('../Mydata/Count_matrix/mm_scATAC_count_matrix_over_all.rownames')]
cols = [i.strip() for i in open('../Mydata/Count_matrix/mm_scATAC_count_matrix_over_all.colnames')]

sc_count = pd.DataFrame(data=count.toarray(),
                        index=idxs,
                        columns=cols)

sc_count.drop(labels=bad_cells.index, axis=1, inplace=True)

# binarise the data
bin_sc_count = sc_count.where(sc_count < 1, 1)
sc_count = sc_count.loc[bin_sc_count.index, bin_sc_count.columns]
bin_sc_count.shape
```

```python
bin_sc_count.head(2)
```

```python
# perform Latent Semantic Indexing Analysis

# get TF-IDF matrix
tfidf = TfidfTransformer(norm='l2', sublinear_tf=True)
normed_count = tfidf.fit_transform(bin_sc_count.T)

# perform SVD on the sparse matrix
lsi = TruncatedSVD(n_components=20, random_state=42)
lsi_r = lsi.fit_transform(normed_count)

lsi.explained_variance_ratio_
```

```python
plate_qc = plate_qc.loc[bin_sc_count.columns]
for i in range(3):
    plate_qc['LSI Dimension {}'.format(i+1)] = lsi_r[:, i]
plate_qc = plate_qc.loc[sorted(plate_qc.index)]
```

```python
# have a look the first dimension and the sequencing depth

g = sns.FacetGrid(plate_qc, hue='Cell', height=6)
g.map(plt.scatter, 'LSI Dimension 1', 'sequencing_depth', s=5)
g.ax.legend(bbox_to_anchor=[1.4, .8])
g.ax.set_yscale('log')
g.ax.set_ylim(1e3, 2e7)
g.ax.set_ylabel('Sequencing depth')

plt.savefig('../Myfigure/LSI_1st_dimension_vs_sequencing_depth.pdf',
            bbox_inches='tight', transparent=True)
```

```python
# have a look the first dimension and the uniq frags

g = sns.FacetGrid(plate_qc, hue='Cell', height=6)
g.map(plt.scatter, 'LSI Dimension 1', 'uniq_frags', s=5)
g.ax.legend(bbox_to_anchor=[1.4, .8])
g.ax.set_yscale('log')
g.ax.set_ylim(1e2, 5e5)
g.ax.set_ylabel('Number of uniqe fragments')

plt.savefig('../Myfigure/LSI_1st_dimension_vs_uniq_frags.pdf',
            bbox_inches='tight', transparent=True)
```

```python
g = sns.FacetGrid(plate_qc, hue='Cell', height=6)
g.map(plt.scatter, 'LSI Dimension 2', 'LSI Dimension 3', s=5)
g.ax.legend(bbox_to_anchor=[1.4, .8])

plt.savefig('../Myfigure/LSI_2st_dimension_vs_LSI_2st_dimension.pdf',
            bbox_inches='tight', transparent=True)
```

```python
# check peaks linked to some marker genes (by annotatePeaks.pl from the HOMER suite)
homer = pd.read_table('../Mydata/homer_annotion_mm_scATAC_peaks.txt', index_col=0)
marker_genes = ['Dux', 'Otx2', 'Nr5a2', 'Rarg', 'Gata4', 'Gata6', 'Klf4', 'Sox2', 'Nanog', 'Pou5f1', 'Cdx2', 'Eomes']

marker_p2g = homer[homer['Gene Name'].isin(marker_genes)]['Gene Name']
marker_count = pd.concat([sc_count, marker_p2g], axis=1, join='inner')
marker_count.head(2)
```

```python
marker_sum = marker_count.groupby('Gene Name').sum()
marker_sum
```

```python
for g in marker_genes:
    plate_qc[g] = marker_sum.loc[g]
```

```python
fig, ax = plt.subplots(figsize=(15, 10), ncols=4, nrows=3)
ax = ax.flatten()

for i, j in enumerate(marker_genes):
    ax[i].scatter(plate_qc['LSI Dimension 2'], plate_qc['LSI Dimension 3'],
                  cmap='Reds', c=plate_qc[j], vmin=0, vmax=10, s=10)
    ax[i].set_title(j, fontsize=12)

plt.tight_layout()

plt.savefig('../Myfigure/Marker_genes.pdf',
            bbox_inches='tight', transparent=True)
```

# A violin plot for uniq_frags and n_counts

```python
# creat the peak matrix as a annotated data
path = '../Mydata/Count_matrix/'
adata = sc.read(path + 'mm_scATAC_count_matrix_over_all.mtx', cache=True).T  # transpose the data
genes = pd.read_csv(path + 'mm_scATAC_count_matrix_over_all.rownames', header=None)
adata.var_names = genes[0]
adata.obs_names = pd.read_csv(path + 'mm_scATAC_count_matrix_over_all.colnames', header=None)[0]
```

```python
# add two columns to adata.obs
adata.obs['n_counts'] = plate_qc['sequencing_depth']
adata.obs['uniq_frags'] = plate_qc['uniq_frags']
```

```python
# using sc.pl.violin to plot, the data must be an adata
sc.pl.violin(adata,['uniq_frags','n_counts'],
             jitter=0.4, multi_panel=True)
```

```python
# change the y scale to log
sc.pl.violin(adata,['uniq_frags','n_counts'],
             jitter=0.4, multi_panel=True, log=True)
```

```python
# another way to change y scale, and this method can change parameters more easy
g = sns.FacetGrid(adata.obs, height=6)
g.map(sns.violinplot,y=adata.obs[['uniq_frags']], order=True)
g.ax.set_ylim(10, 10e6)
g.ax.set_yscale('log')

g.ax.set_ylabel('Number of uniqe fragments')
```
